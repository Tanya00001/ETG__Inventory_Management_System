# Inventory Record

```python
# Initial Inventory record(had run initially, if it will run again, all the entered inventory will be removed)

Inventory = {}
```

## Adding Items into Inventory

```python
# Initially added items

Product_Id   = int(input('Enter Product Id: '))
Name         = input('Enter Product Name: ')
Brand        = input('Enter the Brand of Product: ')
Expire_Date  = input('Enter the Expire date of Product: ')
Price        = int(input('Enter the Price of Product: '))
Weight       = input('Enter the weight of Product: ')
Stock        = int(input('Enter the stock of Product: '))


Inventory[Product_Id] = {'Name' : Name, 'Brand' : Brand, 'Expire Date' : Expire_Date, 'Price(Rs.)' : Price, 'Weight' : Weight, 'Stock' : Stock}


import json
js = json.dumps(Inventory)

fd = open('Inventory.json','w')
fd.write(js)
fd.close()
```

## Reading data from JSON

```python
fd = open('Inventory.json', 'r')

reading = fd.read()

fd.close()
```

```python
# Old Inventory record(had run this cell at the time of adding initial items into the inventory)

reading
```

```python
# Updated Inventory record

reading
```

## Adding new stock

```python
import json
Inventory = json.loads(reading)

Product_Id    = input('Enter Product Id: ')
Stock         = int(input('Enter the New stock: '))
```

```python
# In updated Inventory record, increasing the stock


if(Product_Id in Inventory.keys()):
    print('Product Exist')
    Inventory[Product_Id]['Stock'] = Inventory[Product_Id]['Stock'] + Stock
    
else:
    print('Product does not exist, please enter the following details: \n')
    Name         = input('Enter Product Name: ')
    Brand        = input('Enter the Brand of Product: ')
    Expire_Date  = input('Enter the Expire date of Product: ')
    Price        = int(input('Enter the Price of Product: '))
    Weight       = input('Enter the weight of Product: ')
    Inventory[Product_Id] = {'Name' : Name, 'Brand' : Brand, 'Expire Date' : Expire_Date, 'Price(Rs.)' : Price, 'Weight' : Weight, 'Stock' : Stock}


js = json.dumps(Inventory)

fd = open('Inventory.json','w')
fd.write(js)
fd.close()
```
