```python
# Reading 'Inventory.json' file
# Loading the same as well

import json

fd = open('Inventory.json', 'r')
reading = fd.read()
fd.close()

Inventory = json.loads(reading)
```

```python
# Printing Product Id and Product Name for reference


print('\n\n Product Id       Product Name\n')
print('________________________________')

for i in Inventory.keys():
    print("  "+ i +"       *        "+Inventory[i]['Name'])
    
print('________________________________')     
```

```python
# For checking the complete details of Product
# For checking whether the Product Id exists or not, if not, go for the valid one



i = input('Enter the Product Id for complete details: ')
i = int(i)


if(i>=1001 and i<=1030):
    i = str(i)
    print('\n\n**********Details***********\n')
    print('Product Name :  ', Inventory[i]['Name'])
    print('Brand        :  ', Inventory[i]['Brand'])
    print('Expire Date  :  ', Inventory[i]['Expire Date'])
    print('Price(Rs.)   :  ', Inventory[i]['Price(Rs.)'], 'Rs')
    print('Weight       :  ', Inventory[i]['Weight'])
    print('____________________________')
    
else:
    print('\n\n**********Please enter a valid Product Id**********')
```

```python
# For checking the Total billing amount.


P_Id = input('Enter the Product Id of product you wish to buy: ')
Quan = int(input('Enter the quantity: '))


print('\n\n***************************************')
Cost =  (Inventory[P_Id]['Price(Rs.)'])*Quan
print('\nCost of',Inventory[P_Id]['Name'],'    ', Cost, 'Rs.')


# Discount

if(Cost>=100 and Cost<300):
    dis = ((Cost/100)*5)
    print('\ndiscount        ','   -',round(dis),'Rs.')
elif(Cost>=300 and Cost<500):
    dis = ((Cost/100)*7)
    print('\ndiscount        ','   -',round(dis),'Rs.')
elif(Cost>=500):
    dis = ((Cost/100)*10)
    print('\ndiscount        ','   -',round(dis),'Rs.')   
else:
    dis = 0
    print('\ndiscount        ','   -',round(dis),'Rs.')
    
    
Total = Cost - round(dis)   

print('                  ___________')    
print('Total              ' ,Total, 'Rs.') 
print('                  ___________')
print('\n****************************************')



# As someone buys the product, Stock will be get deducted accordingly and automatically through this and will be updated in 'Inventory.json' 

Inventory[P_Id]['Stock'] = Inventory[P_Id]['Stock'] - Quan

js = json.dumps(Inventory)

fd = open('Inventory.json','w')
fd.write(js)
fd.close()
```

```python
# To make updation on every day basis, it requires to load the 'Sales report.json'


fd = open('Sales report.json','r')
Sales = fd.read()
fd.close()

Sales = json.loads(Sales)
```

```python
# For generating the Transaction Id
T_Id = str(len(Sales)+1)


# Adding Transactions into Sales report by their Transaction Id

Sales[T_Id] = {'Date': Time, 'Product Id' : P_Id, "Customer's Name": C_Name, 'Phone No.': Ph, "Product's Name": Inventory[P_Id]['Name'], 'Price': Inventory[P_Id]['Price(Rs.)'] , 'Quantity': Quan, 'Discount': round(dis), 'Total Amount': Total, 'Payment mode': Pay_mode}
```
